```python
orig_img = cv2.imread('computer-vision-course/data/1dcm.jpg')
orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
img_neg = 255 - orig_img
cv2.imwrite('computer-vision-course/data/img1.jpg', img_neg)
plt.imshow(img_neg)
```

```python
conference_substracted = cv2.subtract(orig_img,img_neg)
plt.imshow(conference_substracted)
```
